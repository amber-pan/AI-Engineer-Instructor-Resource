```python
# adventure_game.py
# A text-based adventure game where the player explores locations and searches for a legendary treasure

def start_game():
    """Start the adventure and introduce the quest"""
    print("\nüåü Welcome to the Ancient Land Adventure! üåü")
    print("A legendary treasure is hidden somewhere in this land.")

    name = input("What is your name, brave explorer? ")
    print(f"\nWelcome, {name}! Your quest begins now.")

    while True:
        print("\nYou stand at a crossroads.")
        print("Do you want to:")
        print("1. Enter the dark forest")
        print("2. Explore the mysterious cave")

        choice = input("> ")

        if choice == "1":
            result = forest_path()
        elif choice == "2":
            result = cave_path()
        else:
            print("Invalid choice. Please choose 1 or 2.")
            continue

        if result == "win":
            print("\nüèÜ Congratulations! You found the legendary treasure!")
            break
        elif result == "lose":
            print("\nüíÄ Your adventure has come to an unfortunate end.")
            if not restart_game():
                break


def forest_path():
    """Forest scenario"""
    print("\nüå≤ You enter the dark forest.")
    print("The trees are tall and the air is thick with mist.")

    print("\nYou see two options:")
    print("1. Follow a quiet river")
    print("2. Climb a tall tree to look around")

    choice = input("> ")

    if choice == "1":
        print("\nYou follow the river and discover a hidden clearing.")
        print("In the center lies a glowing chest.")
        return "win"
    elif choice == "2":
        print("\nYou climb the tree but lose your grip and fall.")
        print("You are injured and cannot continue.")
        return "lose"
    else:
        print("You hesitate too long and get lost in the forest.")
        return "lose"


def cave_path():
    """Cave scenario"""
    print("\nüï≥Ô∏è You step into the mysterious cave.")
    print("It is dark, cold, and you hear strange noises.")

    print("\nYou can:")
    print("1. Light a torch")
    print("2. Proceed carefully in the dark")

    choice = input("> ")

    if choice == "1":
        print("\nThe torch lights up ancient symbols on the walls.")
        print("You follow the symbols to a hidden chamber filled with treasure.")
        return "win"
    elif choice == "2":
        print("\nYou stumble in the darkness and fall into a deep pit.")
        return "lose"
    else:
        print("Frozen by fear, you decide too late.")
        return "lose"


def restart_game():
    """Ask the player if they want to restart"""
    print("\nWould you like to restart the game?")
    print("1. Yes")
    print("2. No")

    choice = input("> ")

    if choice == "1":
        return True
    else:
        print("\nThanks for playing! Farewell, explorer.")
        return False


# Run the game
start_game()
```
